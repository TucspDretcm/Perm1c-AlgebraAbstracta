# Parte 1: Función Inversa

```python
def Euclides(a, b):
  if b == 0:
    return a
  else:
    return Euclides(b, a%b)
```

```python
def Ext_Euclides(a, b):
  if b == 0:
    return (a, 1, 0)
  else:
    d, x_, y_ = Ext_Euclides(b, a%b)
    x, y = y_, x_ - int(a/b)*y_
    return (d, x, y)
```

<h4>Para que se pueda cumplir la inversa debe cumplir que el mcd(a,n)=1 y hacemos o podemos cumplir esa condicion con Euclides(); para hallar el "x" el cual será la inversa de a usamos Euclides Extendido; y finalmente a ese x lo normalizamos metiendolo en el circulo de $Z_{27}$ con x mod n.</h4>

```python
def Inversa(a, n):
  if Euclides(a, n) == 1:
    mgd, x, y = Ext_Euclides(a, n)
    return x % n     # "x mod n"  para que la inversa este entre el circulo de Z_n
  else:
    print("No existe inversa")
```

<img src="https://github.com/TucspDretcm/Perm1c-AlgebraAbstracta/blob/main/imgs/cifradoYdescifrado.JPG?raw=true">

```python
words = list("ABCDEFGHIJKLMNÑOPQRSTUVWXYZ")
words.index("D")   # la posicion de "D" esta en el index 3, words[3] = "D"
```

```python
import random
random.randint(0,26)  # dominio = [0, 26] el cual pertenece al cirulo de Z_27
```

```python
# Generador random de llaves:
a = random.randint(0,26)
while a%3==0:   # validacion donde "a" no puede ser mod de 3 ya que 27 es multiplo de 3 el cual si "a" lo fuera no tendría inversa(a').
  a = random.randint(0,26)
b = random.randint(0,26)

print("Llave Generada: {%d, %d}" % (a,b))
```

```python
def Cifrado(cadena):
  nueva_cadena = ""

  for i in cadena:
    nueva_cadena += words[ (a * words.index(i) + b) % 27]
  
  return nueva_cadena

def Descifrado(cadena):
  a_ = Inversa(a, 27)
  nueva_cadena = ""

  for i in cadena:
    nueva_cadena += words[ (a_ * (words.index(i) - b)) % 27]
  
  return nueva_cadena
```

```python
mensaje_cifrado = Cifrado("SOS")
mensaje_cifrado
```

```python
Descifrado(mensaje_cifrado)
```
